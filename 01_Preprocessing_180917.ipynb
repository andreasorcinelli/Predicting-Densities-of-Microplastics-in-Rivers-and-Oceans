### aggregating ocean plastics and city data

```python
import geopy.distance
```

```python
plastics.head()
```

```python
plastics.shape
```

```python
%%time

# calculating distance between each sample of ocean plastic and the closest 3 cities and population in those cities

df_closest_cities = pd.DataFrame()
master_city_lol = list()
master_closest_cities_plastics = pd.DataFrame()

for idx, row in plastics.iterrows():
    
    p_record = row['Record']
    p_lat = row['Sample_Latitude']
    p_long = row['Sample_Longitude']
    p_pieces = row['Total_Pieces']
    
    for idx, row in cities.iterrows():
        c_lat = row['City_Latitude']
        c_long = row['City_Longitude']
        c_city = row['City']
        c_country = row['Country']
        c_population = row['City_Population']

        specific_city = list()
        specific_city = [#p_lat,p_long,p_pieces,
                         c_city,c_country,c_population,\
                                geopy.distance.great_circle((p_lat,p_long), (c_lat,c_long)).km]

        master_city_lol.append(specific_city)
        #OUT OF CITIES
    
    df_closest_cities = pd.DataFrame(master_city_lol)
    df_closest_cities.head(5)

    df_closest_cities.columns =  ["c_city","c_country","c_population","distance_from_city"] #['p_long','p_lat','p_total_pieces',
    df_closest_cities = df_closest_cities.sort_values(by='distance_from_city').head(3).reset_index(drop=True)
    df_closest_cities_clean = df_closest_cities.stack().to_frame().T
    df_closest_cities_clean

    old_cols = ["c_city","c_country","c_population","distance_from_city"]

    new_cols = list()
    for i in range(1,4):
        for c in old_cols:
            appender = str()
            new_cols.append(c + f"_{i}")
    df_closest_cities_clean.columns = new_cols
    df_closest_cities_clean['p_record']= p_record
    df_closest_cities_clean['p_lat'] = p_lat 
    df_closest_cities_clean['p_long']= p_long
    df_closest_cities_clean['p_total_pieces']= p_pieces
    df_closest_cities_clean.head()
    
    master_closest_cities_plastics = pd.concat([master_closest_cities_plastics,df_closest_cities_clean])
master_closest_cities_plastics.head()
```

```python
master_closest_cities_plastics.shape
```

```python
master_closest_cities_plastics.to_csv('ocean_plastics_cities.csv')
```
